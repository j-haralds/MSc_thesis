```python
import pybamm
import matplotlib.pyplot as plt
```

```python
# load model
model = pybamm.lithium_ion.SPMe()

# create simulation
simulation = pybamm.Simulation(model)

solution = simulation.solve([0, 3600])  # time interval in seconds

c_s_p = solution["Negative particle concentration"]
# Radial coordinate
r = solution["Negative particle radius [m]"]
t = solution["Time [s]"]

plt.plot(r, c_s_p(t[-1], r))

simulation.plot()
```
